```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.mixture import GaussianMixture

# Leer el archivo CSV con el delimitador ;
datos_tabla = pd.read_csv('Tallas.csv', delimiter=';')

# Imprimir las columnas del DataFrame
print(datos_tabla.columns)
datos_tabla['Precio'] = datos_tabla['Precio'].str.replace(',', '').astype(float)
```

```python
print(datos_tabla)
```

```python
caracteristicas = ['Precio', 'Talla', 'Color']
gmm = GaussianMixture(n_components=3)
gmm.fit(datos_tabla[caracteristicas])
etiquetas = gmm.predict(datos_tabla[caracteristicas])
datos_tabla['Etiquetas'] = etiquetas
print(datos_tabla)
```

```python
plt.scatter(datos_tabla['Precio'], datos_tabla['Talla'], c=datos_tabla['Etiquetas'])
plt.xlabel('Precio')
plt.ylabel('Talla')
plt.title('Cl√∫steres Generados por Mezcla Gaussiana')
plt.show()
```
